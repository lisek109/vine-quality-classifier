```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Preparing X and y data
X = df.drop(columns=["is good", "quality"])
y = df["is good"]
```

```python
# Split data into 80/20 trening/test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Train model
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, RocCurveDisplay
import matplotlib.pyplot as plt
```

```python
# To improve models performance I will apply feauters scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
# Evalute models prfomance on tets set using various metrics and showing ROC curve
y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:,1]

print("ACC", accuracy_score(y_test, y_pred))
print("PREC", precision_score(y_test, y_pred))
print("REC", recall_score(y_test, y_pred))
print("F1", f1_score(y_test, y_pred))
print("ROC-AUC", roc_auc_score(y_test, y_proba))

cm = confusion_matrix(y_test, y_pred)
print(cm)

RocCurveDisplay.from_estimator(model, X_test_scaled, y_test)
plt.show()
```

```python
# Precision and recall scores remain low, which suggests that the dataset is imbalanced. I will apply SMOTE and see if this help.
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
# Using smote to create balanced data
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
print("Resampled class distribution:", Counter(y_resampled))
```

```python
# df.kurt() showed that there is many outliers espesielly in chlorides, residual sugar and sulphates. RobustScaler is derfor better for scaling data
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler()
X_resampled = robust_scaler.fit_transform(X_resampled)
```

```python
# Checking the overall standard deviation 
X_resampled.std()
```

```python
X_train_resampled,X_test_resampled,y_train_resampled,y_test_resampled = train_test_split(X_resampled, y_resampled,test_size=0.2, random_state=11)
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)

print("X_train shape:", X_train_resampled.shape)
print("X_test shape:", X_test_resampled.shape)

print("X_train shape:", y_train_resampled.shape)
print("Y_test shape:", y_test_resampled.shape)
```

```python
model_results = []

# Function for training model
def training(model_name, model):
    # Train the model
    model.fit(X_train_resampled, y_train_resampled)
    
    # Prediction
    X_train_resampled_pred = model.predict(X_train_resampled)
    X_test_resampled_pred = model.predict(X_test_resampled)
    
    # Accuracy scores
    accuracy_train = accuracy_score(y_train_resampled, X_train_resampled_pred)
    accuracy_test = accuracy_score(y_test_resampled, X_test_resampled_pred)
    
    # Precision score
    precision_train = precision_score(y_train_resampled, X_train_resampled_pred)
    precision_test = precision_score(y_test_resampled, X_test_resampled_pred)
    
    # Recall score
    recall_train = recall_score(y_train_resampled, X_train_resampled_pred)
    recall_test = recall_score(y_test_resampled, X_test_resampled_pred)
    
    # Report
    report = classification_report(y_test_resampled, X_test_resampled_pred)
    
    # Print results
    print(f"[bold cyan]The accuracy Score for model {model_name} (Train): [/bold cyan] {accuracy_train:.2f}\n")
    print(f"[bold magenta]The Accuracy Score of training by {model_name} (Test):[/bold magenta] {accuracy_test:.2f}\n")
    
    print("[bold yellow]Classification Report:[/bold yellow]")
    print(report)

    # Save results
    model_results.append({
        "Model": model_name,
        "Test Accuracy": accuracy_test,
        "Test Precision:": precision_test,
        "Test_Recall:": recall_test
    })
```

```python
log = LogisticRegression()
training("Logistic Regression", log)
```

```python
from sklearn.tree import DecisionTreeClassifier
dest=DecisionTreeClassifier()
training("Decision Tree Classifier",dest)
```

```python
randf = RandomForestClassifier()
training("Random Forest Classifier",randf)
```

```python
SVC = SVC()
training("SVC",SVC)
```

```python
from sklearn.naive_bayes import GaussianNB
guss = GaussianNB()
training("GaussianNB",guss)
```

```python
results_df = pd.DataFrame(model_results)
print("[bold green]Model Comparison Table:[/bold green]")
print(results_df.sort_values(by="Test Accuracy", ascending = False))
```
