```python
# Creating new column which will be my target "y" variable
df['is good'] = (df['quality'] >= 7).astype(int)
```

```python
# The dataset is imbalanced
df["is good"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Preparing X and y data
X = df.drop(columns=["is good", "quality"])
y = df["is good"]
```

```python
# Split data into 80/20 trening/test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Train model
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, RocCurveDisplay
import matplotlib.pyplot as plt
```

```python
# To improve models performance I will apply feauters scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
# Evalute models prfomance on tets set using various metrics and showing ROC curve
y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:,1]

print("ACC", accuracy_score(y_test, y_pred))
print("PREC", precision_score(y_test, y_pred))
print("REC", recall_score(y_test, y_pred))
print("F1", f1_score(y_test, y_pred))
print("ROC-AUC", roc_auc_score(y_test, y_proba))

cm = confusion_matrix(y_test, y_pred)
print(cm)

RocCurveDisplay.from_estimator(model, X_test_scaled, y_test)
plt.show()
```

```python
# Accuracy is solid 86.9%
# Precision is 0.55 and recall is 0.34- suggest that only 55% of wines predicted as "good" are actually godd- moderate/low precision
# Model misses many truly good wines - low recall
# F1 0.42 - showing an imbalance between precision and recall
# ROC-AUC is 0.85 - means model distinguish good between good and bad wines
```

```python
# Changing threshold and see what if it improve model
y_proba = model.predict_proba(X_test_scaled)[:,1]
y_pred = (y_proba > 0.4).astype(int)

print("ACC", accuracy_score(y_test, y_pred))
print("PREC", precision_score(y_test, y_pred))
print("REC", recall_score(y_test, y_pred))
print("F1", f1_score(y_test, y_pred))
print("ROC-AUC", roc_auc_score(y_test, y_proba))

cm = confusion_matrix(y_test, y_pred)
print(cm)

RocCurveDisplay.from_estimator(model, X_test_scaled, y_test)
plt.show()
```

```python
# It seems that the poor PREC, REC, and F1 scores are caused by imbalanced data.
# I will use class_weight='balanced' and try out 3 different models.
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
```

```python
cross_val_score(LogisticRegression(class_weight='balanced'), X_train_scaled, y_train)
```

```python
cross_val_score(LogisticRegression(class_weight='balanced'), X_train_scaled, y_train, scoring="f1")
```

```python
# SVC model without and with SVC(class_weight='balanced'
```

```python
results = cross_validate(SVC(), X_train_scaled, y_train,
                         scoring=["accuracy", "precision", "recall", "f1"], cv=5)

print("accuracy:", results["test_accuracy"])
print("Precision:", results["test_precision"])
print("Recall:", results["test_recall"])
print("F1:", results["test_f1"])
```

```python
results = cross_validate(SVC(class_weight='balanced'), X_train_scaled, y_train,
                         scoring=["accuracy", "precision", "recall", "f1"], cv=5)

print("accuracy:", results["test_accuracy"])
print("Precision:", results["test_precision"])
print("Recall:", results["test_recall"])
print("F1:", results["test_f1"])
```

```python
# RandomForestClassifier model without and with SVC(class_weight='balanced'
```

```python
results = cross_validate(RandomForestClassifier(), X_train_scaled, y_train,
                         scoring=["accuracy", "precision", "recall", "f1"], cv=5)

print("accuracy:", results["test_accuracy"])
print("Precision:", results["test_precision"])
print("Recall:", results["test_recall"])
print("F1:", results["test_f1"])
```

```python
results = cross_validate(RandomForestClassifier(class_weight='balanced'), X_train_scaled, y_train,
                         scoring=["accuracy", "precision", "recall", "f1"], cv=5)

print("accuracy:", results["test_accuracy"])
print("Precision:", results["test_precision"])
print("Recall:", results["test_recall"])
print("F1:", results["test_f1"])
```
